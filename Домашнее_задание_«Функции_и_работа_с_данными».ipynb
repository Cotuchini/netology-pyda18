**Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

**Решение 1.1**

```python
df2 = pd.read_csv('ratings.csv')

def ocenka(rating):
  if rating >= 4.5:
    return 'Высокий'
  elif 2 < rating <= 4:
    return 'Средний'
  elif rating <= 2:
    return 'Низкий' 

df2['class'] = df2['rating'].apply(ocenka)
df2.head(20)
```

**Решение 1.2**

```python
df2['class']  = df2['rating'].apply(lambda rating: 'Высокий' if (rating >= 4.5) else ('Низкий' if rating <= 2 else 'Средний'))
df2.sample(20)
```

**Задание 2** Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
df1 = pd.read_csv('keywords.csv')
df1.sample(20)

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
df1.sample(10)

def def_reg(word):
  for region, cities in geo_data.items():
    for city in cities:
            if city in word.lower():
                return region
  return 'undefined'        

df1['region'] = df1['keyword'].apply(def_reg)
df1.sample(20)
```
