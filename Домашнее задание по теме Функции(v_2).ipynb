```

documents = [
        {"type": "passport", "number": "2207 876234", "name": "Василий Гупкин"},
        {"type": "invoice", "number": "11-2", "name": "Геннадий Покемонов"},
        {"type": "insurance", "number": "10006", "name": "Аристарх Павлов"}
      ]
 
directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }

#1.информация о владельце 
def owner(numbers):
  for doc_numbers in documents:
    if doc_numbers["number"] == numbers:
      print('Владелец документа: ', doc_numbers["name"])
      break
  else:
    print('Такого номера документа нет в базе.')

#2.на какой полке находится документ    
def find_shelf(numbers):
    res = 'Документ не найден в базе'
    for i in directories.keys():
        if numbers in directories[i]:
          res = 'Документ хранится на полке: ' + i
    print(res)
      
#3.информация о всех документах 
def all_docs():
  k = 1
  for i in documents:   
   print(k, '"'+i['type'], '"'+i['number']+'"', '"'+i['name']+'"')
   k += 1
   find_shelf(i['number'])
   
     
#4.добавление новой полки
def new_shelf(as_shelf):
  if as_shelf in directories.keys():
    print('Такая полка уже существует')
  else:  
    directories.setdefault(as_shelf, [])
    print('Полка добавлена. Текущий перечень полок: ',', '.join(directories.keys()))

#5.удаление полки  
def del_shelf(num_shelf):
 if num_shelf in directories:
    if len(directories[num_shelf]) == 0: 
        del(directories[num_shelf])
        print('Полка удалена. Текущий перечень полок: ',', '.join(directories.keys()))
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ',', '.join(directories.keys()))
 else: print('Такой полки не существует. Текущий перечень полок: ',', '.join(directories.keys()))     

def main():      
 while True:
  command = input('\n \
  Введите одну из команд: p - информация о владельце документа, l - полная информация о документах, \n \
  s - на какой полке находится документ, as - добавить полку, ds - удалить полку. \n \
  Для выхода наберите q. \n \
  Команда: ')
  if command == 'p':
    owner(input('\nВведите номер документа:'))
  elif command == 'l':
    all_docs()
  elif command == 's':
    find_shelf(input('\nВведите номер документа:'))
  elif command == 'as':
    new_shelf(input('\nВведите номер полки:'))
  elif command == 'ds':
    del_shelf(input('Введите номер полки: '))
  elif command == 'q':
    break
  else:
    print('Некорректная команда')

main()
```
